2. How does broadcasting work in NumPy?
  - Broadcasting in NumPy is a powerful feature that allows operations between arrays of different shapes without the need to manually resize them.
  - Instead of throwing an error when shapes don't match, NumPy 'broadcasts' the smaller array across the larger one so that they can work together element-wise.
  - Broadcasting makes code cleaner and more efficient, especially for mathematical and data manipulation tasks.
  - How it works :
      - Alignment : NumPy compares the shapes of the arrays starting from the last dimension and moving backward.
      - Compatibility Rules : Two dimensions are compatible when they are equal or one of them is 1.
      - Stretching : If a dimension is 1, NumPy stretches it to match the other array in that dimension.

```python
# Example of broadcasting
import numpy as np

arr1 = np.zeros((3,4)) # 2D array filled with zeros
arr2 = np.array([1, 2, 3, 4]) # 1D array with 4 elements
result = arr1 + arr2
print(result) # Resulting shape: (3, 4)
```

4. Explain the use of the groupby() method in Pandas.
  - The groupby() method in Pandas is used to group data based on the values in one or more columns and then perform operations on each group independently.
  - It is especially useful for summarizing, aggregating and analyzing large datasets in a structured way.

```python
# Example of groupby() :
import pandas as pd

data = {
    'Employee': ['Rakesh', 'Kirti', 'Rajesh', 'Dhaval', 'Jay'],
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT'],
    'Salary': [50000, 60000, 45000, 47000, 70000]
}

df = pd.DataFrame(data)

# Group by 'Department' and calculate the average salary for each department
result = df.groupby("Department")["Salary"].mean()
print(result)
```

10. What is the significance of hierarchical indexing in Pandas?
  - Hierarchical indexing (also known as MultiIndexing) in Pandas is a feature that allows us to have multiple levels of indexes (row or column labels) in a DataFrame or Series. This makes it easier to work with more complex or multi-dimensional data while still keeping it in a 2D format.
  - Hierarchical indexing gives us more flexibility and structure when working with complex data.
  - It allows us to organize data in layers, like grouping by year and then by month.
  - We can perform operations on higher-level groups without manually filtering.
  - It Makes easier to reshape data with methods like .stack(), .unstack() or .pivot_table().
  - It maintains a clean structure even for complex datasets.

```python
# Example of hierarchical indexing :
import pandas as pd
data = {
    'Marks' : [85, 90, 75, 88]
}

# Create a multi-level index
index = pd.MultiIndex.from_tuples([
    ("Class A", "Dhruv"),
    ("Class A", "Rohit"),
    ("Class B", "Raj"),
    ("Class B", "Rakesh")
], names = ["Class", "Student"])

#Create DataFrame
df = pd.DataFrame(data, index = index)
print(df)
```

17. Explain the difference between apply() and map() in Pandas.
  - The apply() and map() functions in Pandas are both used for applying functions to data.
  - apply() :
      - Used on : Entire DataFrame or Series.
      - Functionality : Allows us to apply a function along a particular axis (rows or columns) of a DataFrame or to each element in a Series.
      - When to use : It is used when we need to apply a function across rows or columns (in a DataFrame).
  - map() :     
      - Used on : Series (not DataFrame).
      - Functionality : Used to map a function to each element in a Series. We can also pass a dictionary, Series or function to map values.
      - When to use : It is used when we need to apply a simple function element-wise to a Series or when we want to replace values based on a mapping (e.g., mapping values to categories).

```python
# Example of apply() :
import pandas as pd
data = {
    "A" : [1, 2, 3],
    "B" : [4, 5, 6]
}

df = pd.DataFrame(data)
# Apply function to each column
result1 = df.apply(lambda x : x.sum())
print("Output of apply() :")
print(result1)

print()
print()

# Example of map() :
s = pd.Series([1,2,3,4])

# Apply function to each element
result2 = s.map(lambda x : x + 1)
print("Output of map() :")
print(result2)
```

19. How does Pandas simplify time series analysis?
  - Pandas makes time series analysis much easier by offering powerful tools and methods that are specifically designed to handle date and time data efficiently.
  - Here's how Pandas simplifies time series tasks :
      -  Datetime indexing : Pandas allows us to use dates as index values in a DataFrame or Series.
      - Date parsing and conversion : We can easily convert strings to datetime objects using pd.to_datetime() (e.g., pd.to_datetime(['2023-01-01', '2023-01-02']))
      - Resampling : Resample data to a different frequency (e.g., daily to monthly) and apply operations like sum, mean, etc. (e.g., data.resample('M').sum()).
      - Time-based slicing : With a datetime index, we can slice data based on date ranges (e.g., data['2023-01-01' : '2023-01-03'])
      - Rolling and expanding windows : Easily compute moving averages and other rolling statistics (e.g., data.rolling(window = 2).mean()).
      - Shifting and lagging : We can shift our data forward or backward in time (e.g., data.shift(1)), useful for comparing past vs. present.
      - Built-in date ranges : Quickly generate custom date ranges (daily, weekly, monthly, etc.) e.g., pd.date_range(start = '2023-01-01', periods = 10, freq = 'W')

20. What is the role of a pivot table in Pandas?
  - The pivot table in Pandas is a powerful tool used to summarize, organize and analyze data in a structured, spreadsheet-like format.
  - Role of pivot table in Pandas :

      - Reshapes datasets - Changes long-form data into a wide format for analysis.
      - Reduces manual work - Automatically groups and calculates values (no need for writing loops).

21. Why is NumPy's array slicing faster than Python's list slicing?
  - The reason NumPy's array slicing is faster than Python's list slicing comes down to how NumPy arrays are stored and managed in memory compared to Python lists.
  - Contiguous Memory Block (C-style) :    
      - NumPy arrays are stored in a single, contiguous block of memory.
      - Python lists are made up of pointers to objects, which are scattered throughout memory.
  - Fixed data type :
      - NumPy arrays hold elements of the same data type, so operations (like slicing) are simpler and faster.
      - Python lists can contain mixed types (e.g., integers, strings), so slicing involves more overhead.
  - Low-Level optimizations :     
      - NumPy is written in C, so slicing is handled using optimized, compiled code.
      - Python lists use high-level, interpreted code, which is inherently slower.
  - Views Instead of Copies :      
      - Slicing a NumPy array returns a view (a lightweight reference to the same data in memory).
      - Slicing a Python list creates a new list (a full copy of the elements).
